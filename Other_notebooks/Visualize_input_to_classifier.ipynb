```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_pickle('dataframes/df_input_to_classifier.pkl')
```

```python
df.head()
```

```python
y = df['annotations']

unique_annotations = set(y)

y = y.replace(['N', 'A'], 'N+A')
y = y.replace(unique_annotations - {'N+A', 'V'}, 'Other')

x = df[['alg1', 'alg3', 'alg4', 'alg5']].astype(int)
x.head()
```

```python
plt.figure(figsize=(6, 6)) # Set the size of the figure

sns.scatterplot(x=x['alg4'], y=x['alg5'], hue=y, style=y, 
                markers=['o', 's', 'D'])

# Adjust legend
plt.legend(title='Klasa', fontsize=12, title_fontsize=14, loc='lower left')

```

```python
data_to_pairplot = df[['alg1', 'alg3', 'alg4', 'alg5']].astype(int)
data_to_pairplot['y'] = y

pairplot = sns.pairplot(data_to_pairplot, vars=['alg1', 'alg3', 'alg4', 'alg5'], hue='y')

# Get the legend object
legend = pairplot._legend
pairplot.fig.set_size_inches(12, 8)  # Set the width and height as needed

# Set the fontsize of the legend
legend.set_title("Klasy", prop={'size': 20})  # Adjust the title size if needed
for text in legend.get_texts():
    text.set_fontsize(20)  # Adjust the font size of the labels



# Show the plot
plt.show()
```

```python

data_to_pairplot = df[['alg1', 'alg3', 'alg4', 'alg5']].astype(int)
data_to_pairplot['y'] = y

# Create a pairplot
pairplot = sns.pairplot(data_to_pairplot, vars=['alg1', 'alg3', 'alg4', 'alg5'], hue='y',
                        markers=['o', 's', 'D'], corner=True)

# Adjust the figure size to make the plot wider
pairplot.fig.set_size_inches(12, 10)  # Set the width and height as needed

# Get the legend object and move it
legend = pairplot._legend


# Show the plot
plt.show()
```
