```python
import Algorithms as alg
import numpy as np
import pandas as pd
import csv


df = pd.read_pickle('dataframes/df_mitdb.pkl')
df_algs = pd.read_pickle('dataframes/df_algorithms.pkl')
df_algs5 = pd.read_pickle('dataframes/df_alg5_turkish.pkl')
```

```python
ecg = df['ECG']
case = 200

W1 = 26200
W2 = 26800
```

```python
df_algs5.head()
```

```python
import matplotlib.pyplot as plt


alg1 = df_algs['Alg1'][case]
alg2 = df_algs['Alg2'][case]
alg3 = df_algs['Alg3'][case]
alg4 = df_algs['Alg4'][case]
alg5 = df_algs5['Alg5'][case]



signal = ecg[case]
time = np.arange(W1, W2)

def mask(arr, W1, W2):
    mask_greater_than = arr > W1
    mask_smaller_than = arr < W2

    # Combine the masks using the logical AND operator
    combined_mask = mask_greater_than & mask_smaller_than

    return arr[combined_mask]


alg1 = mask(alg1, W1, W2)
alg2 = mask(alg2, W1, W2)
alg3 = mask(alg3, W1, W2)
alg4 = mask(alg4, W1, W2)
alg5 = mask(alg5, W1, W2)

plt.figure(figsize=(40, 20))
plt.scatter(alg1, 2*np.ones_like(alg1), color='orange', linewidth=25)
plt.scatter(alg2, 2.5*np.ones_like(alg2), color='orange', linewidth=25)
plt.scatter(alg3, 3*np.ones_like(alg3), color='orange', linewidth=25)
plt.scatter(alg4, 3.5*np.ones_like(alg4), color='orange', linewidth=25)
plt.scatter(alg5, 4*np.ones_like(alg5), color='orange', linewidth=25)

plt.scatter(alg1, 2*np.ones_like(alg1), color='orange', linewidth=15)
plt.scatter(alg2, 2.5*np.ones_like(alg2), color='orange', linewidth=15)
plt.scatter(alg3, 3*np.ones_like(alg3), color='orange', linewidth=15)
plt.scatter(alg4, 3.5*np.ones_like(alg4), color='orange', linewidth=15)
plt.scatter(alg5, 4*np.ones_like(alg5), color='orange', linewidth=15)

plt.scatter(alg1, 2*np.ones_like(alg1), color='orange', linewidth=5)
plt.scatter(alg2, 2.5*np.ones_like(alg2), color='orange', linewidth=5)
plt.scatter(alg3, 3*np.ones_like(alg3), color='orange', linewidth=5)
plt.scatter(alg4, 3.5*np.ones_like(alg4), color='orange', linewidth=5)
plt.scatter(alg5, 4*np.ones_like(alg5), color='orange', linewidth=5)


plt.plot([W1, W2], [2, 2], color='gray')
plt.plot([W1, W2], [2.5, 2.5], color='gray')
plt.plot([W1, W2], [3, 3], color='gray')
plt.plot([W1, W2], [3.5, 3.5], color='gray')
plt.plot([W1, W2], [4, 4], color='gray')

plt.plot(time, signal[W1:W2], color='orange', linewidth=12)

plt.tick_params(axis='both', which='major', labelsize=45)
# # Increase the size of axis labels
# plt.xlabel('Time', fontsize=20)
# plt.ylabel('Signal', fontsize=20)

plt.yticks([-1, -0.5, 0, 0.5, 1, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0], [-1, -0.5, 0, 0.5, 1, 1.5, 'alg1', 'alg2', 'alg3', 'alg4', 'alg5'])


```
